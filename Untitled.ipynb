```python
drug_counts = pd.value_counts(therapy_df['DrugName'], normalize=True)
mask = drug_counts > 0.01
tail_count = drug_counts.loc[~mask].sum()
drug_counts = drug_counts.loc[mask]
drug_counts['Other'] = tail_count
drug_counts.plot(kind='pie', figsize=(16, 16), subplots=True)
plt.ylabel('')
plt.show()
```

```python
def get_drug_table(df):
    table = collections.defaultdict(set)
    for _, row in df.iterrows():
        diagnosis = row['Diagnosis']
        drug_name = row['DrugName']
        table[drug_name].add(diagnosis)
    drug_table = {}
    for k in table.keys():
        drug_table[k] = len(table[k])
    return drug_table
```
