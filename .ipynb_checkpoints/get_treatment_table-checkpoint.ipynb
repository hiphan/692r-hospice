```python
def get_treatment_table(df):
    table = collections.defaultdict(set)
    for _, row in df.iterrows():
        patient_id = row['PatientID']
        diagnosis = row['Diagnosis']
        drug_name = row['DrugName']
        table[patient_id].add((diagnosis, drug_name))
    return table
```

```python
treatment_table = get_treatment_table(therapy_df)
```
